```python
import unicodedata
```

```python
for n,i in enumerate(df_test["home_team"]):
    df_test["home_team"][n] = unicodedata.normalize("NFKC", i)
print(df_test["home_team"])
pd.get_dummies(df_test["home_team"])
```

```python
for n,i in enumerate(df_test["away_team"]):
    df_test["away_team"][n] = unicodedata.normalize("NFKC", i)
pd.get_dummies(df_test["away_team"])
```
